```python
DATA_PATH = "../../data/GSE2034_series_matrix.txt"
```

```python
import pandas as pd
```

```python
with open(DATA_PATH, "r") as f:
    for i in range(20):
        print(f.readline().strip())
```

```python
with open(DATA_PATH, "r") as f:
    for i, line in enumerate(f):
        if line.startswith("\"ID_REF\""):
            print(f"Expression matrix starts at line {i}")
            break
```

```python
dfe = pd.read_csv(
    DATA_PATH,
    sep="\t",
    skiprows=54
)

print(dfe.shape)
```

```python
print(dfe.iloc[:5, :6])
```

```python
dfe = dfe.set_index("ID_REF")
```

```python
dfe.head()
```

```python
DATA_PATH_CL = "../../data/GSE2034_clinical.txt"
```

```python
with open(DATA_PATH_CL, "r") as f:
    for i in range(20):
        print(f.readline().strip())
```

```python
with open(DATA_PATH_CL, "r") as f:
    for i, line in enumerate(f):
        if line.startswith("PID"):
            print(f"Expression matrix starts at line {i}")
            break
```

```python
dfc = pd.read_csv(
    DATA_PATH_CL,
    sep="\t",
    skiprows=7
)

print(dfc.shape)
```

```python
dfc = dfc.rename(columns={
    "GEO_asscession_number": "GEO_accession_number"
})
dfc.head()
```

```python
dfc = dfc.drop('PID', axis=1)
```

```python
dfc.head()
```

```python
print(dfe.info())
```

```python
dfe = dfe.drop(["!series_matrix_table_end"], axis=1)
```

```python
print(dfc.info())
```

```python
dfe_patients = set(dfe.columns)
dfc_patients = set(dfc["GEO_accession_number"])

print(len(dfe_patients))
print(len(dfc_patients))
print(len(dfe_patients & dfc_patients))
```

```python
dfe.index.name = "GEO_accession_number"
dfe = dfe.T
dfe.head()
```

```python
print("Missing count expr",dfe.isnull().sum().sum())
print("Missing count clinical",dfc.isnull().sum().sum())
print("Expression matrix:", dfe.shape)
print("Clinical data:", dfc.shape)
```

```python
data = dfe.merge(
    dfc,
    left_index=True,
    right_on="GEO_accession_number"
)
```

```python
data.head()
```

```python
data.set_index("GEO_accession_number")
```

```python
DATA_PATH_SAVE = "../../data/final/"
```

```python
dfe.to_csv(f"{DATA_PATH_SAVE}/expression_matrix.csv", index=False)
dfc.to_csv(f"{DATA_PATH_SAVE}/clinical_data.csv", index=False)
data.to_csv(f"{DATA_PATH_SAVE}/final_data.csv", index=False)
```
