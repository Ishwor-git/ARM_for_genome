```python
import pandas as pd
from pathlib import Path
```

```python
DATA_MAIN = "../../data/"
DATA_PATH = "final/final_data.csv"
master = pd.read_csv(DATA_MAIN + DATA_PATH)
master.shape
```

```python
# Clinical columns
clinical_cols = [
    "GEO_accession_number",
    "lymph_node_status",
    "time_to_relapse",
    "relapse",
    "ER_Status",
    "Brain_relapses"
]

gene_cols = [
    c for c in master.columns
    if c not in clinical_cols and not c.startswith("AFFX")
]

gene_variance = master[gene_cols].var().sort_values(ascending=False)
```

```python
sizes = [
    50,
    100,
    250,
    500,
    1000,
    2500,
    5000,
    10000,
    len(gene_variance)
]
```

```python
Path(f"{DATA_MAIN}/transactions").mkdir(exist_ok=True)
gene_medians = master[gene_cols].median()
```

```python
for size in sizes:

    selected_genes = gene_variance.index[:size]
    medians = gene_medians[selected_genes]

    filename = f"{DATA_MAIN}/transactions/transactions_{size}.txt"

    with open(filename, "w") as f:

        for _, row in master.iterrows():

            transaction = []

            # Optional: keep patient ID
            transaction.append(row["GEO_accession_number"])

            for gene in selected_genes:

                if row[gene] >= medians[gene]:
                    transaction.append(f"{gene}_HIGH")
                else:
                    transaction.append(f"{gene}_LOW")

            transaction.append(f"RELAPSE_{row['relapse']}")

            f.write(",".join(map(str, transaction)) + "\n")

    print(f"Saved {filename}")
```
